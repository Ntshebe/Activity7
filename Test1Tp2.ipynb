**OUTLIER**

1. Detecting Outliers
 * IQR Method: Outliers are defined as data points that
 lie below the lower bound or above the upper bound, calculated as:
   *  Lower Bound = Q1 - 1.5 * IQR
   *  Upper Bound = Q3 + 1.5 * IQR
 *  Quartiles ( and ) divide the dataset into four parts, and IQR measures the spread between the middle 50% of values.

2. Removing Outliers:
   *   Rows with values outside these bounds for the
       selected column are removed from the dataset.
   

3.  Displaying Results:
   *  The cleaned dataset excludes rows with outliers
      for Accommodation cost and Transportation cost.

**Data Visualization**

```python
gendercount = df['Traveler gender'].value_counts()
plt.pie(gendercount,labels=gendercount.index, autopct='%1.1f%%')
plt.title("Traveler Gender Distribution")
plt.show()
```

```python

df['Accommodation cost'] = df['Accommodation cost'].replace('[^\d.]', '', regex=True).astype(float)
df['Transportation cost'] = df['Transportation cost'].replace('[^\d.]', '', regex=True).astype(float)
df['Traveler name'] = df['Traveler name'].astype(str)

df_sliced = df.iloc[0:10]

plt.figure(figsize=(10, 6))
plt.plot(df_sliced['Traveler name'], df_sliced['Accommodation cost'], label='Accommodation Cost', color='blue', marker='o', linestyle='-')
plt.plot(df_sliced['Traveler name'], df_sliced['Transportation cost'], label='Transportation Cost', color='red', marker='s', linestyle='--')
plt.xlabel("Traveler Name")
plt.ylabel("Cost ($)")
plt.title("Traveler Name vs Accommodation & Transportation Costs (First 10 Travelers)")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
df_sliced = df.iloc[0:10]
plt.figure(figsize=(10, 6))

plt.scatter(df_sliced['Traveler name'], df_sliced['Accommodation cost'], label='Accommodation Cost', color='blue', marker='o', linestyle='-')
plt.scatter(df_sliced['Traveler name'], df_sliced['Transportation cost'], label='Transportation Cost', color='red', marker='s', linestyle='--')
plt.xlabel("Traveler Name")
plt.ylabel("Cost ($)")
plt.title("Traveler Name vs Accommodation & Transportation Costs (First 10 Travelers)")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
